```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
```

```python
#Datos de cr√©ditos personales
Creditos= pd.read_csv("/Bank_Personal_Loan.csv")
Creditos.tail()

Creditos.head()

```

```python
#Busco si hay alguna relaci√≥n entre el ingreso y el promedio de gastos en tarjetas de cr√©dito
Creditos.plot(kind="scatter",x="Income",y="CCAvg")
plt.title("Ingresos vs gastos")
plt.show()
```

```python
s# Datos de entrenamiento
X_train, X_test, y_train, y_test = train_test_split(Creditos.Income, Creditos.CCAvg)
```

```python
#Datos de entrenamiento y prueba
plt.scatter(X_train, y_train, label="Datos de entrenamiento", color="red", alpha =.7)
plt.scatter(X_test, y_test, label="Datos de prueba", color="blue", alpha =.7)
plt.legend()
plt.title("Split de datos de prueba")

plt.show()
```

```python
#Modelo l√≠neal
LR = LinearRegression()
LR.fit(X_train.values.reshape(-1,1), y_train.values)
```

```python
#
prediction = LR.predict(X_test.values.reshape(-1,1))

#L√≠nea de predicci√≥n
plt.plot(X_test, prediction, label = "Regresi√≥n Lineal", color="red")
plt.scatter(X_test, y_test, label="Datos de prueba", color="blue", alpha =.7)
plt.legend()
plt.title("Regresi√≥n Lineal")
plt.show()
```

```python
# Predicci√≥n del CCAvg, con base en un ingreso proporcionado

LR.predict([[75]])
```

```python
#Score
LR.score(X_test.values.reshape(-1,1), y_test.values)
```
