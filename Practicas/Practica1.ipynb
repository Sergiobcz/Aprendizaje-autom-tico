```python
import matplotlib.pyplot as plt
from matplotlib import style
import seaborn as sns

import statsmodels.api as sm
from scipy import stats
```

```python
print(deps[149])
sueldo=uanl.loc[deps[149]]["Sueldo Neto"]
#Histograma + curva normal teórica
# ==============================================================================

# Valores de la media (mu) y desviación típica (sigma) de los datos
mu, sigma = stats.norm.fit(sueldo)

# Valores teóricos de la normal en el rango observado
x_hat = np.linspace(min(sueldo), max(sueldo), num=100)
y_hat = stats.norm.pdf(x_hat, mu, sigma)

# Gráfico
fig, ax = plt.subplots(figsize=(7,4))
ax.plot(x_hat, y_hat, linewidth=2, label='normal')
ax.hist(x=sueldo, density=True, bins=30, color="#3182bd", alpha=0.5)
ax.plot(sueldo, np.full_like(sueldo, 0), '|k', markeredgewidth=1)
ax.set_title('Distribución de sueldo de SECRETARÍA DE IGUALDAD E INCLUSIÓN')
ax.set_xlabel('sueldo')
ax.set_ylabel('Densidad de probabilidad')
ax.legend();
```
